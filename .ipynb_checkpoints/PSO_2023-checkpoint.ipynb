# Particle Swarm Optimization (PSO)


## Introduction 

Particle Swarm Optimization or PSO is one of the population based optimization techniques in which a collection of particles (swarm) or agents is launched in a high dimensional parameters space. Following their own exploration and that of others the swarm is able to find the global maximum resonably quickly. 

## History 
Originally PSO was proposed by Kennedy and Eberhart in 1995 and since then many variants of PSO have been proposed either to fix the shortcomings of original PSO or to add new capabilities or deal with new chllanges. 


## Main components 
Irrespective to which of the variants of PSO you follow, all of them share the following common features.

- Particles : Particles or set of computational agents 
- Cost function f(X) : The function for which we want to find the glbal minimum/maximum. 
- Search Space X: Multidimeninsinal search space 
- Search Boundary (Xmin, Xmax) 
- Personal Best or Pbest : The set of points at which the particles have found their best values. 
- Global Best or Gbest : The pbest amoung all pbests at which the function has minimum/maximum value. 

## Main considerations 

- Number of particles 
- Initial Condtion 
  - How to set the initial position
  - How to set the initial velocities
- Boundary conditions : What to do when particles reach the boundary 
-  Convergence Criteria 



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed (seed=102)

def parabola (X):
    return np.sum (np.square(X))



def func_val (x, func):
     y = [ func (x[i,:])  for i in range (0, len(x))]
     return np.array(y) 
    

def plot_func (xmin, xmax, func):
    
     x = np.linspace (xmin,xmax,100)
     y = np.linspace (xmin,xmax,100)
    
     X, Y = np.meshgrid (x, y)
     Z = np.zeros ([len(x), len(y)])   
     for i in range (0, len(x)):
         for j in range (0, len(y)):
            z = np.array ([x[i],y[j]])
            Z[i,j] =  func (z)
    
    
     # Creating figure
     fig = plt.figure(figsize =(14, 9))
     ax = plt.axes(projection ='3d')
 
     # Creating plot
     ax.plot_surface(X, Y, Z)
  
     # show plot
     plt.title("Optimization function [Parabola]")
     plt.show()

                
        
        
class PSO:
    def __init__(self, func, xmin, xmax):
        self.npart = 10
        self.ndim = len (xmin)
        self.xmin = xmin 
        self.xmax = xmax
        
        self.pos = np.ndarray ([self.npart, self.ndim])
        self.vel = np.ndarray ([self.npart, self.ndim])
       
        self.pbest = np.ndarray ([self.npart])
        self.xpbest = np.ndarray ([self.npart, self.ndim])
        
        self.gbest  = 1.0E10 
        self.xgbest = np.ndarray ([self.ndim])
        #self.update_best()     


        self.w = 0.7
        self.c1 = 2.19
        self.c2 = 2.19
        
        print("pos:",self.pos.size, self.npart, self.ndim,self.pos.shape)
        
        
    def set_inc(self):
        
     
        r1 = np.random.random ([self.npart, self.ndim])
        r2 = np.random.random ([self.npart, self.ndim])
       
        size = self.xmax - self.xmin 
    
        self.vmax = 0.0025 * size 
        self.vmin = - self.vmax  
    
        for i in range (0, self.npart):
            for j in range (0, self.ndim):
                self.pos[i,j] = self.xmin[j] + r1[i, j] * size[j]
                self.vel[i,j] = (0.5 - r2[i,j]) * 0.025 * size [j]
    
   
        pval = func_val (self.pos, parabola)
        self.pbest = pval
        self.xpbest =  self.pos 
        
        self.update_best()
     
        
    def boundary (self, ):
        for i in range (0, self.npart):
            for j in range (0, self.ndim):
                if self.pos[i,j] > self.xmax[j]:
                    self.pos[i,j] = self.xmax[j]
                    self.vel[i,j] *=-1.0
                if self.pos[i,j] < self.xmin [j]:
                    self.pos[i,j] = self.xmin[j]
                    self.vel[i,j] *=-1.0 
   
    def update_best (self):
        y  = func_val (self.pos, parabola)
        print("y=",np.min(y))
        for i in range (0, self.npart):
             if y[i] <= self.pbest[i]:
                 self.pbest[i] =  y[i].copy()
                 self.xpbest[i,:] = self.pos[i,:].copy()
             if self.pbest[i] <= self.gbest:
                self.gbest = self.pbest[i]
                self.xgbest = self.xpbest[i,:]
        

        #print("ppp",self.gbest, self.xgbest, self.gbest)


    def evolve  (self):
        r1 = np.random.random ([self.npart,self.ndim])
        r2 = np.random.random ([self.npart,self.ndim])   
 
        for i in range (0, self.npart):
             for j in range (0, self.ndim):
                y1  =  self.w * self.vel[i,j] 
                y2  =  self.c1 * r1[i,j] * (self.xpbest[i,j] - self.pos[i,j])
                y3  =  self.c2 * r2[i,j] * (self.xgbest[j] - self.pos[i,j]) 
                    
                self.vel[i,j] = self.vel[i,j] + y1+y2+y3 
                    
                self.vel[i,j] = np.max ([self.vel[i,j], self.vmax[j]])
                self.vel[i,j] = np.min ([self.vel[i,j], self.vmin[j]])
                
                self.pos[i,j] = self.pos[i,j] + self.vel[i,j]  
     
                
        #self.boundary() 
        self.update_best () 
   
    
def plot_inc(P):
    plt.scatter(P.pos[:,0],P.pos[:,1],label='Initial Positions')
    plt.scatter(P.vel[:,0],P.vel[:,1],label='Final posistions')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.show()

        
if __name__ == "__main__":
    plot_func (-10,10,parabola)
    
 
    xmin = [-10,-10]
    xmax = [10, 10] 
    
    
    P = PSO(parabola,np.array(xmin), np.array(xmax))  
  
    P.set_inc()
  
    df = pd.DataFrame(columns=['iter','xgbest','ygbest','gbest'])
    df.loc[0] = [0, P.xgbest[0], P.xgbest[1], P.gbest]

    for i in range (0, 50):
        P.evolve ()
        df.loc[i+1] = [i+1, P.xgbest[0], P.xgbest[1], P.gbest]
     
    print(df)
    
  
    fig, axs = plt.subplots (3,1, figsize=(12,12))
    axs[0].plot (df['iter'], df['xgbest'],'-o')
    axs[1].plot(df['iter'],df['ygbest'],'-o')
    axs[2].plot(df['iter'],df['gbest'],'-o')
    
    axs[0].set_ylabel("x")
    axs[1].set_ylabel("y")
    axs[2].set_ylabel("func")
    axs[2].set_xlabel("num_iterations")
    
    plt.show()
    
```
