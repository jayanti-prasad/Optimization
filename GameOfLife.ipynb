```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

m,n=20,20


fig, ax = plt.subplots(1,1,figsize=(12,12))
ims = []

data = np.random.randint (0,2,[m,n])
print(data)

def bc (i, n):
    if i >= n:
        i = i -n
    if i < 0:
        i = i + n
    return i     
        

for i in range(100):
    new_data = data.copy()
    for j in range (0, m):
        for k in range (0, n):
            num_black = 0
            num_white = 0
            for j1 in range (-1,2):
                for k1 in range (-1,2):
                    if j1 + k1 != 0:
                        if data[bc(j+j1,m),bc(k+k1,n)] == 0:
                            num_black +=1
                        else:
                            num_white +=1

            if num_black in [3,4,5]:
                if data[j,k] == 0:
                    new_data[j,k] = 1
              
            if num_white in [2,3,4]:
                if data[j,k] == 1:
                    new_data[j,k] = 0
              
           
            
            
    data = new_data.copy()
    im = ax.imshow (data, animated=True, cmap='gray')
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)

ani.save('test.mp4', writer = 'ffmpeg', fps = 10)

plt.show()
```
